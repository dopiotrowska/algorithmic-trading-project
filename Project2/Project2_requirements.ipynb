# Lecture 4.4 -Project nr 2

## Project: Combining Signals in Financial Markets with Real-World Testing

---

### **Objective**
- Develop multiple trading signals that, when combined, outperform the markets across various environments.

---

### **Requirements**

#### **1. Signal Development and Market Assumptions**
- Develop multiple signals (e.g., indicators, price patterns, moving averages, sentiment data) and combine them using methods such as:
  - PCA (Principal Component Analysis)
  - Random Forest
  - Other machine learning techniques (Lectures 1, 2, 4).
- Alternatively, take a completely different approach:
  - Identify a statistical market feature and exploit it (simplicity is key).
- Minimize the difference between out of sample and in sample performance of your strategy
- Use **BackTrader** for backtesting (Lecture 3)
- Simulate real-world trading conditions:
  - **Slippage:** 0.2%.
  - **Commission:** 0.4%.
- Implement risk management techniques such as:
  - Stop-loss.
  - Take-profit.
  - Trailing stop (justify the choice of these mechanisms).
- Short selling is allowed (for simplicity -> no liquidity constraints)
- Assume the markets behave similarly to stock market (no margin account, margin calls - for simplicity)
- Trading rules:
  - Open a maximum of **10 positions at once**.
  - Allocate **10% of capital per position**.

---

#### **2. Testing**
- Test the signals across different timeframes.
- Measure performance using the following metrics:
  - **Return Rate.**
  - **Maximum Drawdown.**
  - **Sharpe Ratio (rate =0.03).**
- **Important Deadline:**  
  - By **27th January**, five different time series (1,000 ohlc price entries each) will be uploaded. (1h, 4h, 1d, 1m timespans of the data)
  - Datasets are going to be random (with strong trend, week trend, horizontal move, down trend, very volatile etc) 
  - Test your best strategy on these datasets.
  - Provide tabularized results for them in a form
    - timeseries1, return in %, max dd, sharp ratio

---

#### **3. Performance Evaluation**
- Grades will be awarded based on:
  - Strategy performance.
  - Justification for the effectiveness of the strategy.
- Results will be compared using:
  - **Sharpe Ratio.**
- Additional metrics to consider:
  - **Maximum Drawdown.**
  - **Win Rate.**
